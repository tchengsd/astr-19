```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(size=1000)

width = 0.01
histmin = 0
histmax = 1.01

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,color='green',edgecolor='black')
plt.xlabel('x')
plt.ylabel('N per bin')
plt.savefig('Random.pdf')
```
