```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)

def sin(theta):
    return np.sin(theta)
def cos(theta):
    return np.cos(theta)
```

```python
f, xarr = plt.subplots(1,2)

xarr[0].plot(x, sin(x))
xarr[0].set_xlabel('x')
xarr[0].set_ylabel('sin x')
xarr[0].set_title(r'$\sin x$')

xarr[1].plot(x, cos(x))
xarr[1].set_xlabel('x')
xarr[1].set_ylabel('cos x')
xarr[1].set_title(r'$\cos x$')

f.subplots_adjust(wspace=0.5)

plt.savefig('SineAndCosine.pdf')
```
