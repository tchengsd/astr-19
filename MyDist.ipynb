```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.chisquare(1, size=100)

histmin = np.floor(min(x))
histmax = np.floor(max(x))
width = (histmax - histmin) / 100
histmax += width

bins = np.arange(histmin, histmax, width)
plt.hist(x, bins=bins, alpha=0.5, color='#006947', edgecolor='black')
plt.xlabel('X')
plt.ylabel('# per bin')
plt.savefig('MyDist.pdf')
```
